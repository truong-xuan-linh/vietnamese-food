#**Đổi tên các ảnh và chia train, test**

##Import libraries

```python
import cv2
import glob
import os
```

##Đổi tên ảnh và chia train, test

* Các ảnh được đổi tên theo định dạng "ten_mon_an_sothutu.jpg"
* Với mỗi món ăn, ta lấy 20% ảnh mỗi món làm test và 80% còn lại để train

```python
directory = '/content/drive/MyDrive/CS114/predict_food/data_new'
train_path = '/content/drive/MyDrive/CS114/predict_food/data_split/data_train/'
test_path = '/content/drive/MyDrive/CS114/predict_food/data_split/data_test/'
```

```python
os.chdir(directory)
food = os.listdir()
for f in food:
  path = glob.glob(directory + f + '/*')
  k = 0
  split = int(len(path)*0.2)
  for p in path:
    k=k+1
    if k <= split:
      img = cv2.imread(p)
      cv2.imwrite(test_path + f +"/" + f+ "_"+ str(k) +".jpg", img)
    else:
      img = cv2.imread(p)
      cv2.imwrite(train_path + f +"/" + f+ "_"+ str(k) +".jpg", img)
```

```python
class_names_visual = os.listdir(test_path)
print(os.listdir(test_path + class_names_visual[0]))
```

#**Display data**

##Import libraries

```python
import matplotlib.pyplot as plt
import random
```

##Display multi data

```python
  def display_data(cols, rows, test_path):
    fig = plt.figure(figsize = (20,14))
    i = 0
    for name in class_names_visual:
      image = plt.imread(test_path + '/' + name +'/' + name +'_'+ str(random.randint(1,44)) + '.jpg')
      fig.add_subplot(rows, cols, i + 1)
      plt.imshow(image)
      plt.axis('off')
      plt.title(name, fontsize = 15)
      i = i+1
```

```python
display_data(cols = 6, rows = 5, test_path = test_path)
```
